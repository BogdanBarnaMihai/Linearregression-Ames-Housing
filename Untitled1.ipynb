***Writing My Own Train-Test Split***

Yeah I know sklearn has train_test_split built in, but I wanted to write one myself just to understand what's happening under the hood. It takes the data, shuffles the indices with a fixed random seed so things stay reproducible, then carves off 20% for testing. It works with both DataFrames and numpy arrays, whichever you throw at it. Nothing fancy, but it does the job.



```python
def train_test_split(x,y,test_size=0.2,random_state=42):
  np.random.seed(random_state)
  indices=np.arange(len(x))
  np.random.shuffle(indices)
  if isinstance(y,(pd.Series,pd.DataFrame)):
    y=y.values
  split_indx=int(len(x)*(1-test_size))
  train_index=indices[:split_indx]
  test_index=indices[split_indx:]

  if isinstance(x,(pd.Series,pd.DataFrame)):
    x_train,x_test=x.iloc[train_index],x.iloc[test_index]
  else:
    x_train,x_test=x[train_index],x[test_index]

  y_train,y_test=y[train_index],y[test_index]

  return x_train,x_test,y_train,y_test
```

***PCA From Scratch***

This one took a minute to get right. Instead of importing PCA from sklearn, I built it step by step. It standardizes the training data first—subtract the mean, divide by standard deviation. Then it computes the covariance matrix and runs an eigendecomposition to find the principal components. The eigenvalues get sorted from largest to smallest so we know which directions matter most.

The function figures out how many components to keep based on whatever variance threshold you give it. If you want 95% of the information from the original data, it'll tell you how many components that takes. It prints out the number it picked and the actual variance captured. The same transformation then gets applied to the test set using the parameters learned from training.



[PCA Full Math](https://github.com/BogdanBarnaMihai/Linearregression-Ames-Housing/tree/main/math/PCA)

```python
def PCA(x_train,x_test,threshold=0.95, max_components=None):
  if isinstance(x_train,pd.DataFrame):
    x_train=x_train.values
  if isinstance(x_test,pd.DataFrame):
    x_test=x_test.values
  train_mean=np.mean(x_train,axis=0)
  train_std=np.std(x_train,axis=0)
  train_std[train_std<1e-10]=1

  x_train_std=(x_train-train_mean)/train_std

  cov_matrix=np.cov(x_train_std,rowvar=False)
  eigenvalues,eigenvectors=np.linalg.eigh(cov_matrix)

  idx=eigenvalues.argsort()[::-1]
  eigenvalues=eigenvalues[idx]
  eigenvectors=eigenvectors[:,idx]

  eigenvalues=np.maximum(eigenvalues,0)

  var_exp=eigenvalues/np.sum(eigenvalues)
  cum_var_exp=np.cumsum(var_exp)
  k=np.argmax(cum_var_exp>=threshold)+1

  if max_components is not None:
        k = min(k, max_components)

  print(f"Selected {k} components")
  print(f"Variance explained: {cum_var_exp[k-1]:.2%}")

  k=min(k,min(x_train.shape[0], x_train.shape[1]))

  x_train_pca=x_train_std@eigenvectors[:,:k]

  if x_test is not None:
    x_test_std = (x_test - train_mean) / train_std
    x_test_pca = x_test_std @ eigenvectors[:, :k]
    return x_train_pca, x_test_pca, k, eigenvalues[:k], cum_var_exp[:k]

  return x_train_pca, k, eigenvalues[:k], cum_var_exp[:k]
```

***Trying Different Variance Thresholds***

This is where I test out a few different cutoffs—80%, 85%, 90%, and 95% variance explained. For each one, I run PCA and see how many components we end up with. The results are pretty interesting. At 80% variance we only need 128 components. By the time we hit 95%, we're up to 196. That's a decent reduction from the original number of features but not a dramatic one, which tells me the data doesn't have a ton of redundancy.

The code saves all the results and then picks the best threshold—the one that captured the most variance overall. That transformed data is what we'll use going forward for whatever model comes next.



```python
thresholds = [0.80, 0.85, 0.90, 0.95]
results = {}

for i in thresholds:
  print(f"Threshold: {i}")

  x_train_pca, x_test_pca, k, evals, cumvar = PCA(
      x_train,x_test,threshold=i, max_components=None
  )

  results[i]={
      'x_train': x_train_pca,
      'x_test': x_test_pca,
      'k': k,

      'evals': evals,
      'cumvar': cumvar
  }

best_threshold=max(results,key=lambda x:results[x]['cumvar'][-1])
x_train=results[best_threshold]['x_train']
x_test=results[best_threshold]['x_test']
k=results[best_threshold]['k']
```

***Building a Standard Scaler From Scratch***

Again, I could just import `StandardScaler` from sklearn and call it a day, but where's the fun in that? I wanted to understand exactly what happens when you "standardize" data, so I built my own version.

The basic idea is dead simple: for every column, subtract the mean and divide by the standard deviation. That's it. Now all your features are centered around zero and have a spread of roughly one. It's a crucial step before PCA because we don't want columns with huge ranges (like square footage in the thousands) dominating columns with tiny ranges (like number of bathrooms).

***The Constructor***

I set this up with a few options. You can turn off mean centering or standard deviation scaling if you want—though I'm not sure why you would. There's also a tiny epsilon value to prevent division by zero if a column happens to have no variance at all. The attributes like `mean_` and `std_` start off as None and get filled in later when you actually fit the scaler to data.

***The Fit Method***

This is where the scaler learns the parameters from your training data. It accepts DataFrames, numpy arrays, or really anything that can be coerced into an array. It figures out how many features we're dealing with, computes the mean of each column if mean centering is turned on, and calculates standard deviations if scaling is enabled. If any standard deviation comes out smaller than our epsilon threshold, I set it to 1.0 so we don't blow up later with division errors. Once all that's done, I flip a little flag saying the scaler has been fitted.

***The Transform Method***

Applying the transformation is straightforward. The method checks if we've actually fitted the thing first—no point trying to standardize with parameters that don't exist yet. Then it does the same array conversion dance as the fit method, subtracts the stored means, and divides by the stored standard deviations. Whatever comes out is your standardized data.

***Fit Transform Shortcut***

Just a convenience method that chains fit and transform together. Saves a line of code when you're doing everything in one go on the training set. Nothing clever, just cleaner.

[Standard Scaler](https://github.com/BogdanBarnaMihai/Linearregression-Ames-Housing/tree/main/math/Standard_scaler)


```python
class StandardScaler:
  def __init__(self,mean=True,std=True,eps=1e-10):
    self.mean=mean
    self.std=std
    self.eps=eps
    self.mean_=None
    self.std_=None
    self.n_features=None
    self.is_fitted=None

  def fit(self, x):
    if isinstance (x,pd.DataFrame):
      x=x.values
    elif not isinstance(x, np.ndarray):
      x = np.array(x)
    n_samples,n_features=x.shape
    self.n_features=n_features
    if self.mean:
      self.mean_=np.mean(x,axis=0)
    else:
      self.mean_ = np.zeros(n_features)
    if self.std:
      self.std_=np.std(x,axis=0)
      self.std_[self.std_<self.eps]=1.0
    else:
      self.std_=np.ones(n_features)
    self.is_fitted =True
    return self

  def transform(self,x):
    if not self.is_fitted:
      print("call .fit() first")
    else:
      if isinstance(x,pd.DataFrame):
        x=x.values
      elif not isinstance(x, np.ndarray):
            x = np.array(x)
      return (x - self.mean_) / self.std_

  def fit_transform(self,x):
    self.fit(x)
    return self.transform(x)
```

***Linear Regression with Maximum Likelihood Estimation***

This class implements ordinary least squares regression but approaches it through the lens of maximum likelihood estimation. Instead of just minimizing squared error, we're actually fitting a probabilistic model that assumes the errors follow a normal distribution. This gives us some nice extras—log likelihood, AIC, and BIC—that tell us more than just how well the line fits.

What Gets Stored

The class keeps track of a few things after fitting. Theta holds our regression coefficients. Sigma squared is the estimated variance of the error term. We also compute and store the log likelihood of the data under our fitted model, plus the Akaike and Bayesian information criteria. Those last two are useful for model comparison later.

***The Fit Method***

This does the heavy lifting. First it wrangles the input into numpy arrays regardless of whether you passed in DataFrames, Series, or plain lists. The y vector gets flattened to one dimension so matrix operations behave.

The normal equation gives us our coefficients: theta equals the inverse of X transpose X, multiplied by X transpose y. I wrote this with a small safety net. If the matrix happens to be singular and the straightforward solve fails, it falls back to the pseudo-inverse instead of crashing. Not ideal from a numerical stability standpoint but it keeps things running.

After we have theta, we compute the residuals by subtracting predicted values from actual y. The error variance sigma squared is just the sum of squared residuals divided by sample size—the maximum likelihood estimator without Bessel's correction.

The log likelihood formula for a normal distribution gets evaluated next. It's not pretty but it's standard: a constant term involving pi, a term for the variance, and the sum of squared residuals scaled by twice the variance.

Finally we tack on AIC and BIC. Both penalize model complexity, with BIC being harsher for larger sample sizes since it multiplies the parameter count by the log of n rather than just two.

***Making Predictions***

Dead simple. Check that the model has been fitted, convert whatever input we got into a numpy array, and multiply by theta. That's all linear regression prediction ever is.

***Mean Squared Error***

Just a convenience method that calls predict and computes the average squared difference between predictions and actual values. Useful for a quick performance check on training data, though you'd normally want this on a held-out test set.

***R Squared***

Another quick diagnostic. This calculates the proportion of variance in y that our model explains. It's one minus the ratio of residual sum of squares to total sum of squares. Values close to one mean the model captures most of the variation. Values near zero mean we might as well be guessing the mean every time.



[Linear Regression and MLE Full Math](https://github.com/BogdanBarnaMihai/Linearregression-Ames-Housing/tree/main/math/Linear_Regression%2BMLE)

```python
class LinearRegressionandMle:
  def __init__ (self):
    self.theta=None
    self.sigma=None
    self.log_likelihood=None
    self.aic=None
    self.bic=None
  def fit(self,x,y):
    if isinstance(x,(pd.DataFrame,pd.Series)):
      x=x.values
    elif not isinstance(x,np.ndarray):
      x=np.array(x)
    if isinstance(y,(pd.DataFrame,pd.Series)):
      y=y.values.ravel()
    elif not isinstance(y,np.ndarray):
      y=np.array(y)

    n,d=x.shape
    x = np.hstack([np.ones((n, 1)), x])

    try:
      self.theta=np.linalg.solve(x.T @ x,x.T@y)
    except np.linalg.LinAlgError:
      self.theta = np.linalg.pinv(x.T @ x) @ x.T @ y

    r=y-x@self.theta
    self.sigma2=(r@r)/n

    self.log_likelihood=(-n/2*np.log(2*np.pi)-n/2*np.log(self.sigma2)-(r@r)/(2*self.sigma2))
    k=d+1

    self.aic = -2 * self.log_likelihood + 2 * k
    self.bic = -2 * self.log_likelihood + k * np.log(n)
    return self

  def predict(self,x):
    if self.theta is None:
            raise ValueError("Model not fitted. Call fit() first.")

    if isinstance(x, pd.DataFrame):
        x = x.values
    elif not isinstance(x, np.ndarray):
        x = np.array(x)

    x = np.hstack([np.ones((len(x), 1)), x])

    return x@self.theta

  def mse(self,x,y):
    y_pred=self.predict(x)
    return np.mean ((y_pred-y)**2)

  def r_squared(self,x,y):
    y_pred = self.predict(x)
    ss_res = np.sum((y - y_pred) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    return 1 - (ss_res / ss_tot)
```

***Ridge Regression From Scratch***

Ridge regression is basically linear regression with a penalty on coefficient size. It's what you reach for when your predictors are correlated and ordinary least squares starts misbehaving—coefficients blowing up, high variance, the usual multicollinearity headaches. The penalty term shrinks everything toward zero just enough to keep things stable.

***A Quick Note on the Code***

There's a small typo in the class definition up there. The method is called rideRegression instead of RidgeRegression and the __init__ method is missing the lam parameter in its signature. It should be def __init__(self, lam=1.0): to actually accept and store the regularization strength. I'll write this description assuming those fixes are in place.

***The Constructor***

The class takes one parameter: lambda, which controls how hard we penalize large coefficients. Higher lambda means more shrinkage, lower lambda gets us closer to ordinary least squares. We also set up placeholders for the weight vector and something called effective degrees of freedom, which is a concept from ridge regression that measures model complexity in a penalized world.

***The Fit Method***

The math here is almost identical to ordinary least squares except we add a ridge term to the matrix before solving. Instead of solving X transpose X times w equals X transpose y, we solve (X transpose X plus lambda times identity) times w equals X transpose y. That little lambda times identity matrix is what stabilizes everything.

The identity matrix we add is size d by d, where d is the number of features. Same safety net as before: if the direct solver fails, fall back to the pseudo-inverse.

After solving, the weights are stored and we're ready to predict.

***Making Predictions***

Nothing special here. Multiply the feature matrix by the weight vector. Same as ordinary least squares. The difference is entirely in how those weights were estimated.

***Mean Squared Error and R Squared***

These two helper methods do exactly what you'd expect. They take a feature matrix and the true target values, generate predictions, and compute either the average squared error or the R squared statistic. Useful for evaluating how well the regularization is working on held-out data.

[Ridge Regression and R-Squared Full Math](https://github.com/BogdanBarnaMihai/Linearregression-Ames-Housing/tree/main/math/Ridge_Regression%2BR_Squared)

```python
class rideRegression():
  def __init__ (self,lam=1.0):
    self.lam=lam
    self.w=None
    self.effective_df=None

  def fit(self,x,y):
    if isinstance(x,(pd.DataFrame,pd.Series)):
      x=x.values
    elif not isinstance(x,np.ndarray):
      x=np.array(x)

    if isinstance(y,(pd.DataFrame,pd.Series)):
      y=y.values.ravel()
    elif not isinstance(y,np.ndarray):
      y=np.array(y)

    n,d=x.shape

    try:
      self.w=np.linalg.solve(x.T@x+self.lam*np.eye(d),x.T@y)
    except np.linalg.LinAlgError:
      self.w=np.linalg.pinv(x.T@x+self.lam*np.eye(d))@x.T@y

    return self

  def predict(self,x):
    if self.w is None:
      print("call .fit() first")
    else:
      if isinstance(x,pd.DataFrame):
        x=x.values
      elif not isinstance(x,np.ndarray):
        x=np.array(x)
      return x@self.w

  def mse(self,x,y):
    y_pred=self.predict(x)
    return np.mean((y_pred-y)**2)

  def r_squared(self,x,y):
    y_pred=self.predict(x)
    ss_res=np.sum((y_pred-y)**2)
    ss_tot=np.sum((y-np.mean(y))**2)
    return 1-(ss_res/ss_tot)
```

***Lasso Regression with Coordinate Descent Style Updates***

Lasso is the spicier cousin of ridge regression. Instead of just shrinking coefficients toward zero, it can push them all the way to exactly zero. That means it does feature selection and regularization at the same time. The trick is using an L1 penalty instead of L2, which creates a diamond-shaped constraint region where corners touch the axes—and corners are where coefficients die.

***Constructor Setup***

The class takes a few parameters. Lambda controls the penalty strength like in ridge, but the effect is more dramatic. Learning rate and tolerance govern how the gradient-style updates behave. I'm also setting the maximum iterations to 2000, though hopefully we converge long before that. We're using the StandardScaler from earlier to normalize features, which is absolutely crucial for lasso since the penalty treats all coefficients equally. Without scaling, features with larger magnitude would get penalized unfairly.

The class keeps track of the scaled weights, the weights in original feature space, an intercept term, and how many iterations we actually used before stopping.

***The Soft Thresholding Operator***

This little helper function is the heart of lasso. It takes a value and a threshold, and if the absolute value is below the threshold, it returns zero. Otherwise it shrinks the absolute value by the threshold amount and keeps the original sign. This is what creates sparsity. Every weight update passes through this function, which means small weights get crushed to nothing while larger ones survive with a penalty.

***The Fit Method***

This implements a proximal gradient descent approach rather than coordinate descent, but the idea is similar. Features get standardized first so everything is on the same scale. We initialize all weights to zero and start iterating.

Each iteration computes the gradient of the squared error loss with respect to the current weights. We take a small step in the negative gradient direction, then immediately apply the soft thresholding operator. The threshold is lambda times the learning rate—the product matters because the gradient step size affects how much penalty gets applied.

After the update we check how much the weight vector changed since the previous iteration. If the change drops below our tolerance threshold, we call it converged and break early. This saves time when the solution stabilizes quickly.

Once the scaled weights are found, we transform them back to the original feature space by dividing by the standard deviations from the scaler. The intercept gets computed as the mean difference between actual y and predictions using the unscaled weights. This two-step approach handles the fact that we standardized for fitting but want predictions in the original units.

***Making Predictions***

Predictions apply the same standardization to new data, multiply by the scaled weights, and add the intercept. The intercept captures the overall mean level while the scaled weights handle the relationships with individual features.

***Model Evaluation Helpers***

Mean squared error and R squared work exactly like in the other regression classes. Nothing new there.

***Sparsity Diagnostics***

These last two methods are lasso-specific and honestly pretty useful. n_nonzero counts how many coefficients survived the thresholding—anything with absolute value above a tiny epsilon counts as nonzero. get_sparsity turns that into a percentage, telling you what fraction of features got completely zeroed out. Higher sparsity means a simpler, more interpretable model. This is the whole reason you'd pick lasso over ridge in the first place.



[Lasso Full Math](https://github.com/BogdanBarnaMihai/Linearregression-Ames-Housing/tree/main/math/Lasso_Regressionn)

```python
class lassoegression():
  def __init__(self,lam=1.0,lr=0.01,tol=1e-4,iter=2000):
    self.lam=lam
    self.iter=iter
    self.lr=lr
    self.tol=tol
    self.w = None
    self.w_original = None
    self.intercept_ = None
    self.scaler = StandardScaler()
    self.n_iter_used = 0

  def soft_threshold(self,x,lambda_val):
    return np.sign(x)*np.maximum(np.abs(x)-lambda_val,0)

  def fit(self,x,y,verbose=False):
    if isinstance(x,(pd.DataFrame,pd.Series)):
      x=x.values
    elif not isinstance(x,np.ndarray):
      x = np.array(x)

    if isinstance(y,(pd.DataFrame,pd.Series)):
        y = y.values.ravel()
    elif not isinstance(y,np.ndarray):
        y = np.array(y).ravel()


    x_scaled=self.scaler.fit_transform(x)
    n,d=x_scaled.shape

    self.w=np.zeros(d)
    prev_w=self.w.copy()

    converged = False
    for i in range(self.iter):
      residuals=x_scaled@self.w-y
      grad=(x_scaled.T@residuals)/n
      w_temp=self.w-self.lr*grad
      self.w=self.soft_threshold(w_temp,self.lam*self.lr)

      weight_change=np.linalg.norm(self.w-prev_w)
      if weight_change<self.tol:
        self.n_iter_used=i+1
        converged = True
        break

      prev_w=self.w.copy()

    if not converged:
      self.n_iter_used = self.iter

    self.w_original = self.w / self.scaler.std_
    self.intercept_ = np.mean(y - x @ self.w_original)

    return self

  def predict(self,x):
    if self.w is None:
      print("call .fit() first")
    else:
      if isinstance(x, pd.DataFrame):
          x = x.values
      elif not isinstance(x, np.ndarray):
          x = np.array(x)

      x_scaled = self.scaler.transform(x)
      return x_scaled @ self.w + self.intercept_

  def mse(self,x,y):
    y_pred=self.predict(x)
    return np.mean((y_pred-y)**2)

  def r_squared(self,x,y):
    y_pred=self.predict(x)
    rss=np.sum((y_pred-y)**2)
    tss=np.sum((y-np.mean(y))**2)
    return 1 - (rss / tss)

  def n_nonzero(self):
    if self.w is None:
        return 0
    return np.sum(np.abs(self.w) > 1e-4)

  def get_sparsity(self):
    if self.w is None:
        return 0.0
    n_total = len(self.w)
    n_nonzero = self.n_nonzero()
    return 1.0 - (n_nonzero / n_total)
```

***Finding the Right Amount of Regularization***

Now that both ridge and lasso are implemented, the obvious question is: how much penalty should we actually use? The lambda parameter controls everything, and picking the wrong value means either under-regularizing and getting unstable coefficients or over-regularizing and washing out real signal. We need to test a range and see what works best.

***Testing Ridge Across Lambda Values***

I picked a set of lambda values that spans several orders of magnitude—from barely any regularization at 0.1 all the way up to extremely heavy shrinkage at 1000. For each one, I fit a ridge model on the training data and recorded the mean squared error on both training and test sets.

The loop is straightforward: create the ridge object with the current lambda, fit it, compute both MSEs, and store them. After trying all the values, the best lambda is simply whichever one gave the lowest test error. That's the number we actually care about. Training error will always look better with less regularization, but test error tells us how well the model generalizes.

I also converted the best test MSE to RMSE and put a dollar sign in front of it because hey, we're predicting house prices. Seeing the error in actual dollars is way more intuitive than squared dollars.

***Doing the Same Thing for Lasso***

Lasso gets a similar treatment but with a slightly different range of lambdas. Lasso tends to zero out coefficients more aggressively, so the useful lambda range is shifted. Values that would be moderate for ridge might completely wipe out lasso coefficients. I capped things at 100 instead of 1000.

The lasso loop also tracks something extra: the number of nonzero coefficients at each lambda value. This is where lasso gets interesting. As lambda increases, the model should use fewer and fewer features. Tracking this alongside the MSE helps us understand the tradeoff between simplicity and accuracy.

I also bumped up the maximum iterations to 3000 for lasso since the soft thresholding updates can take longer to settle, especially with larger lambda values where many coefficients are fighting to stay alive.

***What the Numbers Actually Say***

Printing out the best lambda and corresponding RMSE for both methods gives us a direct comparison. Ridge will probably have a lower test error since it uses all available features in a stabilized way. Lasso might have slightly higher error but with the benefit of a much simpler model—potentially throwing out dozens or even hundreds of features entirely.

The dollar sign in the RMSE print statement is a nice touch. If the best model has an RMSE of say twenty thousand dollars, that means our predictions are typically off by about that much. For a housing dataset where prices range from fifty thousand to over seven hundred thousand, that might be pretty reasonable.

***What Comes Next***

With the best lambda values identified, we could dig into the actual coefficients, see which features lasso decided to keep, or compare the two models more thoroughly. But for now, we have tuned versions of both regularized regression approaches ready to go.



```python
lambdas = [0.1, 1.0, 10.0, 50.0, 100.0, 500.0, 1000.0]
ridge_train_mse=[]
ridge_test_mse=[]

for lam in lambdas:
  ridge=rideRegression(lam=lam)
  ridge.fit(x_train,y_train)
  ridge_train_mse.append(ridge.mse(x_train,y_train))
  ridge_test_mse.append(ridge.mse(x_test,y_test))

best_ridge_lambda=lambdas[np.argmin(ridge_test_mse)]
best_ridge_mse=min(ridge_test_mse)

laso_lambdas=[0.1, 0.5, 1.0, 5.0, 10.0, 50.0, 100.0]
lasso_train_mse=[]
lasso_test_mse=[]
lasso_nonzero = []

for lam in laso_lambdas:
  lasso=lassoegression(lam=lam,lr=0.01,iter=3000)
  lasso.fit(x_train,y_train)
  lasso_train_mse.append(lasso.mse(x_train,y_train))
  lasso_test_mse.append(lasso.mse(x_test,y_test))
  lasso_nonzero.append(lasso.n_nonzero())

best_lasso_lam = laso_lambdas[np.argmin(lasso_test_mse)]
best_lasso_mse = min(lasso_test_mse)

print(f"Best Ridge: lambda={best_ridge_lambda}, Test MSE={best_ridge_mse:,.0f}, RMSE=${np.sqrt(best_ridge_mse):,.0f}")
print(f"Best Lasso: lambda={best_lasso_lam}, Test MSE={best_lasso_mse:,.0f}, RMSE=${np.sqrt(best_lasso_mse):,.0f}")
```

```python
def create_poly(x_train,x_test,degree=2):
  if degree==1:
    scaler=StandardScaler()
    return scaler.fit_transform(x_train),scaler.fit_transform(x_test),scaler
  x_train_poly=x_train.copy()
  x_test_poly=x_test.copy()

  for d in range(2,degree+1):
    x_train_poly=np.hstack((x_train_poly,x_train**d))
    x_test_poly=np.hstack((x_test_poly,x_test**d))

  scaler=StandardScaler()
  x_train_poly=scaler.fit_transform(x_train_poly)
  x_test_poly=scaler.fit_transform(x_test_poly)

  return x_train_poly,x_test_poly,scaler
```

```python
print("Creating polynomial features (degree=2)...")
x_train_poly, x_test_poly, poly_scaler = create_poly(x_train, x_test, degree=2)
print(f"Original features: {x_train.shape[1]}")
print(f"Polynomial features: {x_train_poly.shape[1]}")
```

```python
ols_poly=LinearRegressionandMle()
ols_poly.fit(x_train_poly,y_train)
ols_poly_mse=ols_poly.mse(x_test_poly,y_test)

ridge_poly=rideRegression(lam=best_ridge_lambda)
ridge_poly.fit(x_train_poly,y_train)
ridge_poly_mse=ridge_poly.mse(x_test_poly,y_test)

lasso_poly=lassoegression(lam=best_lasso_lam,lr=0.005,iter=5000)
lasso_poly.fit(x_train_poly,y_train)
lasso_poly_mse=lasso_poly.mse(x_test_poly,y_test)

ols_test=LinearRegressionandMle()
ols_test.fit(x_test,y_test)
ols_test_mse=ols_test.mse(x_test,y_test)

print(f"OLS (poly):   Test MSE: {ols_poly_mse:,.0f}  |  RMSE: ${np.sqrt(ols_poly_mse):,.0f}")
print(f"Ridge (poly): Test MSE: {ridge_poly_mse:,.0f}  |  RMSE: ${np.sqrt(ridge_poly_mse):,.0f}")
print(f"Lasso (poly): Test MSE: {lasso_poly_mse:,.0f}  |  RMSE: ${np.sqrt(lasso_poly_mse):,.0f}")
print(f"Non-zero features: {lasso_poly.n_nonzero()}/{x_train_poly.shape[1]}")
```

```python
all_models = {
    'OLS (Linear)': ols_test_mse,
    'Ridge (Linear)': best_ridge_mse,
    'Lasso (Linear)': best_lasso_mse,
    'OLS (Poly deg=2)': ols_poly_mse,
    'Ridge (Poly deg=2)': ridge_poly_mse,
    'Lasso (Poly deg=2)': lasso_poly_mse
}

print("="*70)
print("FINAL MODEL COMPARISON")
print("="*70)
print(f"{'Model':<25} {'Test MSE':<20} {'Test RMSE':<15} {'Improvement':<15}")
print("-"*70)

baseline_mse = all_models['OLS (Linear)']

for name, mse in sorted(all_models.items(), key=lambda x: x[1]):
    rmse = np.sqrt(mse)
    improvement = (1 - mse/baseline_mse) * 100
    print(f"{name:<25} {mse:>15,.0f}      ${rmse:>10,.0f}      {improvement:>5.1f}%")

winner_name = min(all_models, key=all_models.get)
winner_mse = all_models[winner_name]
winner_rmse = np.sqrt(winner_mse)

print("\n" + "="*70)
print(f"WINNER: {winner_name}")
print(f"Test RMSE: ${winner_rmse:,.0f}")
print(f"Predictions off by ~${winner_rmse:,.0f} on average")
print(f"Target mean: ${y.mean():,.0f}, Target std: ${y.std():,.0f}")
print(f"RMSE/Mean ratio: {winner_rmse/y.mean():.1%}")
print("="*70)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()

models_to_plot = [
    ('OLS (Linear)', LinearRegressionandMle(), x_train, y_train, x_test, y_test),
    ('Ridge (Linear)', rideRegression(lam=best_ridge_lambda), x_train, y_train, x_test, y_test),
    ('Lasso (Linear)', lassoegression(lam=best_lasso_lam, lr=0.01, iter=3000), x_train, y_train, x_test, y_test),
    ('OLS (Poly deg=2)', LinearRegressionandMle(), x_train_poly, y_train, x_test_poly, y_test),
    ('Ridge (Poly deg=2)', rideRegression(lam=best_ridge_lambda), x_train_poly, y_train, x_test_poly, y_test),
    ('Lasso (Poly deg=2)', lassoegression(lam=best_lasso_lam, lr=0.005, iter=5000), x_train_poly, y_train, x_test_poly, y_test)
]

for idx, (name, model, X_tr, y_tr, X_te, y_te) in enumerate(models_to_plot):
    if name in ['OLS (Linear)', 'OLS (Poly deg=2)']:
        model.fit(X_tr, y_tr)
    elif name == 'Ridge (Linear)':
        model = rideRegression(lam=best_ridge_lambda)
        model.fit(X_tr, y_tr)
    elif name == 'Lasso (Linear)':
        model = lassoegression(lam=best_lasso_lam, lr=0.01, iter=3000)
        model.fit(X_tr, y_tr)
    elif name == 'Ridge (Poly deg=2)':
        model = rideRegression(lam=best_ridge_lambda)
        model.fit(X_tr, y_tr)
    elif name == 'Lasso (Poly deg=2)':
        model = lassoegression(lam=best_lasso_lam, lr=0.005, iter=5000)
        model.fit(X_tr, y_tr)

    y_pred = model.predict(X_te)

    mse = np.mean((y_pred - y_te) ** 2)
    rmse = np.sqrt(mse)
    r2 = 1 - np.sum((y_te - y_pred) ** 2) / np.sum((y_te - np.mean(y_te)) ** 2)

    ax = axes[idx]
    ax.scatter(y_te, y_pred, alpha=0.5, edgecolors='k', linewidth=0.5)

    min_val = min(y_te.min(), y_pred.min())
    max_val = max(y_te.max(), y_pred.max())
    ax.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Perfect Prediction')

    ax.set_xlabel('Actual Price ($)', fontsize=10)
    ax.set_ylabel('Predicted Price ($)', fontsize=10)
    ax.set_title(f'{name}\nRMSE: ${rmse:,.0f} | R²: {r2:.3f}', fontsize=11, fontweight='bold')
    ax.legend(loc='upper left')
    ax.grid(True, alpha=0.3)

    ax.ticklabel_format(style='plain', axis='both')
    ax.set_xticklabels([f'${int(x):,}' for x in ax.get_xticks()])
    ax.set_yticklabels([f'${int(y):,}' for y in ax.get_yticks()])

plt.suptitle('Actual vs Predicted House Prices - Model Comparison', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
